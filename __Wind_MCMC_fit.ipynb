```python
from _Wind_fit_MCMC_2D_Boundary_Layer import *
from _worm_dataloader import *
```

# D1

```python
# path
datapath = ('E:/_Charge_Worm_PNAS_2025/Data/worm_wind_data/data_D1.xls')
t, data, vec = loaddata(datapath, fps = 1e4, show = True, truncate_front = 0, truncate_end = 0)
```

```python
# ODE class
# scales = [a_w, u_x, u_y, U]
init_guess = [100e-6, vec[0], vec[1], 0.2, 5e-3]
scales = [1e-5, 1e-1, 1e-1, 1e-1, 1e-3]

worm = wind_fit_joint(t, data, [data[0,0],data[0,1]], init_guess, scales, True, RW_sigma = 5e-3,
              para_var = 1e1, burn_in = 15000)

# Sample
paras_init_5 = np.array((0,0,0,0,0))[None,:]
worm.sampler.sample(paras_init_5, nIter = 30000)
```

```python
worm.show_iter2()
worm.show_mean_sol(show = True, long_t = 0)
worm.show_coeff(show_PDF = False, view = [30,-60])
```
