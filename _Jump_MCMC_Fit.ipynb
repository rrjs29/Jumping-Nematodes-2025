```python
from _Jump_MCMC_Fit_2D import *
from _worm_dataloader import *
```

# A01 videos

```python
# path
datapath = ('E:/_Charge_Worm_PNAS_2025/Data/worm_no_charge_data/data_A01.xls')
t, data, vec = loaddata(datapath, fps = 1e4, show = True, truncate_front = 0, truncate_end = 150)
```

```python
# ODE class
# scales = [a_w, u_x, u_y]
init_guess = [100e-6, vec[0], vec[1]]
scales = [1e-5, 1e-1, 1e-1]

worm = Jump_fit_joint(t, data, [data[0,0],data[0,1]], init_guess, scales, RW_sigma = 10e-3,
              para_var = 1e1, burn_in = 5000)

# Sample
paras_init_3 = np.array((0,0,0))[None,:]
worm.sampler.sample(paras_init_3, nIter = 10000)
```

```python
worm.show_iter2()
worm.show_mean_sol(show = True, long_t = 0)
worm.show_coeff(show_PDF = False, view = [30,-60])
```
