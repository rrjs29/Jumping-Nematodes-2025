# B700V 02

```python
# path
datapath = ('E:/_Charge_Worm_PNAS_2025/Data/worm_fly_data/data_B700V_02.xls')
t, data, vec = loaddata(datapath, fps = 1e4, show = True)
```

```python
# ODE class
# scales = [a_w, Q_w, u_x, u_y]

init_guess = [100e-6, 1e-13, vec[0], vec[1]]
scales = [10e-6, 1e-12, 1e-1, 1e-1]

worm = charge_fit_joint(t, data, [data[0,0],data[0,1]], init_guess, scales, RW_sigma = 5e-4,
              para_var = 1e1, burn_in = 50000, a_f = 1e-3, phi_f = 700)

# Sample
paras_init_4 = np.array((0,0,0,0))[None,:]
worm.sampler.sample(paras_init_4, nIter = 100000)
```

```python
worm.show_iter2()
worm.show_mean_sol(show = True, long_t = 0)
worm.show_coeff(show_PDF = False, view = [30,-60])
```
